```python
#Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import 311 NYC service request data from csv
Service_request_df=pd.read_csv('311_Service_Requests_from_2010_to_Present.csv')
```

```python
#status of tickets
Service_request_df['Status'].value_counts().plot(kind='barh',alpha=0.6,figsize=(7,7))
plt.show()
```

```python
#Explore the data for noise related compalints call counts over city 
Noise_complaint_df=Service_request_df[Service_request_df['Complaint Type'].str.contains("Noise")]
Noise_complaint_df.head()
```

```python
Noise_complaint_df['Complaint Type'].value_counts()
```

```python
# convert the created date column to datetime 
Noise_complaint_df['Created Date'] = pd.to_datetime(Noise_complaint_df['Created Date'])
```

```python
# creating month and year for our two datasets so we can groupby month in our analysis
Noise_complaint_df['month'] = Noise_complaint_df['Created Date'].dt.month
```

```python
# updated dataframe with month and year column
print (Noise_complaint_df.shape)
Noise_complaint_df.columns
```

```python
# create new df to isolate complaint counts over time# create 
month_noise_complaint_df = Noise_complaint_df[["month", "Complaint Type", "Unique Key"]]
```

```python
# group by year to see count of unstacked over time
Group_month_noise_complaint_df = month_noise_complaint_df.groupby(['month', 'Complaint Type'])['Unique Key'].count().unstack('Complaint Type')
```

```python
fig, ax = plt.subplots(figsize=(20,12)) 
Group_month_noise_complaint_df.plot(kind='bar', stacked=True, ax=ax)
plt.title('Noise Complaints by Month',fontsize=30)
plt.ylabel('Call Count',fontsize=15)
plt.xlabel('Month',fontsize=15)

plt.legend(loc=1, prop={'size': 10})
plt.show
```

```python
#Display the complaint type and city together
fig, ax = plt.subplots(figsize=(25,30)) 
Service_request_df.groupby(['Complaint Type', 'City'])['Unique Key'].count().unstack().plot(kind='bar', stacked=True,ax=ax)

plt.title('Compalint call counts over city',fontsize=30)
plt.ylabel('Count' ,fontsize=15)
plt.xlabel('Complaint Types' ,fontsize=15)

plt.legend(loc=1, prop={'size': 15})
plt.show
```

```python
#Find major complaint types
Droped_NA_complaints_type_df=Service_request_df.dropna(subset=['Complaint Type'])

DroppedNA_DF=pd.DataFrame({"count":
                           Droped_NA_complaints_type_df.groupby(['Complaint Type','City']).size()}).reset_index()

DroppedNA_DF
```

# Major complaint types and their count

```python
#Display the major complaint types and their count
Major_complaints_df=Droped_NA_complaints_type_df.groupby('Complaint Type')
Sorted_Major_complaints_df = Major_complaints_df.size().sort_values(ascending=False).to_frame('count').reset_index()
Sorted_Major_complaints_df
```

```python
#top 10 complaint types 
Sorted_Major_complaints_df.head(10)
```

```python
#Complaint type with bar plot 
Major_complaints_Sorted_df=Major_complaints_df.size().sort_values(ascending=False)
Major_complaints_Sorted_df.plot(kind='barh',alpha=0.6,figsize=(10,20))
plt.title('Count vs. Complaint Types ')
plt.xlabel('Count')
plt.ylabel('Complaint Types')
plt.show()
```

```python
#top 10 complaintstype

Major_complaints_Sorted_df=Major_complaints_df.size().sort_values(ascending=False)
Top10_ComplaintTypes_DF =Major_complaints_Sorted_df.head(10)

Top10_ComplaintTypes_DF.plot(kind='bar',alpha=0.6,figsize=(10,20))
plt.title('Count vs. Complaint Types ')
plt.ylabel('Count')
plt.xlabel('Complaint Types')
plt.show()
```
